```python
import requests

res = requests.get("http://ggoreb.com/dinosaur/info.jsp")
```

```python
result = res.json() #문자를 List와 Dictionary로 변경

for re in result: #for-each
    print(re.get("kind"))
    
print("\n")
    
for index in range(len(result)):
    print(result[index].get("kind"))
```

```python
for value in result[0]["dinosaurs"][0]["imgs"]:
    print(value)
```

```python
for re in result:
    kind = re.get("kind")#re["kind"]
    print(kind)
```

```python
# result => 전체 리스트 [{}, {},]

for dinosaur in result:
    dinosaurs = dinosaur.get("dinosaurs") #전체리스트에서 dinosaurs 리스트만 추출
    for dino in dinosaurs:
        title = dino.get("title")# dinosaurs 리스트에서 title 리스트만 추출

        print(title)
```

```python
import requests

res = requests.get("http://ggoreb.com/api/shop.jsp")
```

```python
result = res.json() # 문자를 List와 Dictionary로 변경
```

```python
shop = result.get("list")

for item in shop: #result["list"]:
    lng = item.get("lng")
    lat = item.get("lat")
    print("위도: " + str(lng) + "   경도: " + str(lat) )
```
